```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Read the data from the Excel sheet
df = pd.read_excel("United_States_Healthcare_Spending.xlsx") 

#size up figure of chart
plt.figure(figsize=(10, 5))

#set plotlines for graph
plt.plot(df['Year'], df['Total Healthcare Spending (Billions USD)'], label='Total Healthcare Spending (Billions USD)')
plt.plot(df['Year'], df['Hospital Care (Billions USD)'], label='Hospital Care (Billions USD)')
plt.plot(df['Year'], df['Physician Services (Billions USD)'], label='Physician Services (Billions USD)')
plt.plot(df['Year'], df['Prescription Drugs (Billions USD)'], label='Prescription Drugs (Billions USD)')
plt.plot(df['Year'], df['Nursing Care (Billions USD)'], label='Nursing Care (Billions USD)')
plt.plot(df['Year'], df['Other Healthcare Services (Billions USD)'], label='Other Healthcare Services (Billions USD)')

#extend y axis
plt.ylim(0, 5000)

#place and format legend
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')


#label x and y axises
plt.xlabel('Year')
plt.ylabel('USD in Billions')
plt.title('Growth of Healthcare Spending in United States since 1995')

plt.show()


df = pd.read_excel("United_States_BioTech_Spending.xlsx") 

df.plot(kind="bar", x='Year', y='BioTech Spending (Total)',color="red")

# Create bar graph

plt.xlabel('Year')
plt.ylabel('BioTech Spending (Total)')
plt.title('United States BioTech Spending since 2000')



plt.show()

#show plot

```

The R&D industry increased dramatically since 2000, with percentage of overall spending in terms of U.S. GDP from 2.6 in 2000 to 3.5 in 2021.
